```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaled_market_data = StandardScaler().fit_transform(df_market_data)

# Review the first five rows of the `scaled_market_data` DataFrame
scaled_market_data[:5]
```

```python
# Create a DataFrame with the scaled data and set coinid column as index
scaled_market_data_df = pd.DataFrame(
    scaled_market_data,
    columns=df_market_data.columns,
    index=df_market_data.index
)

# Display first 5 rows
scaled_market_data_df.head(5)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(scaled_market_data_df)
    inertia.append(km.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_plot = elbow_df.hvplot.line(
    x="k",
    y="inertia",
    title="Elbow Curve",
    xticks=k
)
elbow_plot
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** Looking at the Elbow Curve above, the best value for k is 4 due to the sharp decrease in slope for k values greater than 4.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
k_model = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the scaled data
k_model.fit(scaled_market_data_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
k_clusters = k_model.predict(scaled_market_data_df)

# Print the resulting array of cluster values.
print(k_clusters)
```

```python
# Create a copy of the DataFrame
df_market_data_predicted = df_market_data.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_market_data_predicted['market_segments'] = k_clusters

# Display sample data
df_market_data_predicted.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

market_data_predicted_plot = df_market_data_predicted.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="market_segments",
    hover_cols=["coin_id"],
    title="Market Segments",
)
market_data_predicted_plot
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
pca_market_data = pca.fit_transform(scaled_market_data_df)

# View the first five rows of the DataFrame. 
pca_market_data[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

```python
# Calculate the total explained variance of the three principal components.
print(round(sum(pca.explained_variance_ratio_), 3))
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance of the three principal components is 89.5%

```python
# Creating a DataFrame with the PCA data and set the coinid column as index
pca_market_data_df = pd.DataFrame(
    pca_market_data,
    columns=["PC1", "PC2", "PC3"],
    index=df_market_data.index
)

# Display sample data
pca_market_data_df.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_pca = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia_pca = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `pca_market_data_df`
# 3. Append the model.inertia_ to the inertia list
for i in k_pca:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(pca_market_data_df)
    inertia_pca.append(km.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data_pca = {"k": k_pca, "inertia": inertia_pca}
# Create a DataFrame with the data to plot the Elbow curve
elbow_data_pca_df = pd.DataFrame(elbow_data_pca)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_pca_plot  = elbow_data_pca_df.hvplot.line(x="k", y="inertia", title="PCA Elbow Curve", xticks=k_pca)
elbow_pca_plot
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** Looking at the Elbow Curve above, the best value for k is 4 due to the sharp decrease in slope for k values greater than 4.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No, it does not differ from the best k value found using the original data.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
k_model_pca = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the PCA data
k_model_pca.fit(pca_market_data_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
k_clusters_pca = k_model_pca.predict(pca_market_data_df)
# Print the resulting array of cluster values.
print(k_clusters_pca)
```

```python
# Create a copy of the DataFrame with the PCA data
df_market_data_predicted_pca = pca_market_data_df.copy()

# Add a new column to the DataFrame with the predicted clusters
df_market_data_predicted_pca["pca_market_segments"] = k_clusters_pca

# Display sample data
df_market_data_predicted_pca.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
market_data_pca_predicted_plot = df_market_data_predicted_pca.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="pca_market_segments",
    hover_cols=["coin_id"],
    title="PCA Market Segments"
)
market_data_pca_predicted_plot
```
